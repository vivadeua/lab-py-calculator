# Practice | Make a Python calculator



1. Write a program that asks the user for two input numbers (a) and (b). Both integers.
2. Add both numbers (+) and print results
3. Subtract both numbers (-) and print results
3. Multiply both numbers (*) and print results
4. Divides both numbers (/) and print results

#### Please note that you are highly encouraged to enhance your calculators with additional features such as handling exceptions (e.g., division by zero).

```python
## TODO : Place your code here
a = int(input())
b = int(input())
action = int(input("""Choose what we will do:
Enter 1 for Addition
Enter 2 for Substraction
Enter 3 for Multiplikation
Enter 4 for Division"""))
if action == 1:
    print (a + b)
elif action == 2:
    print (a - b)
elif action == 3:
    print (a * b)
elif action == 4:
    try:
        print (a / b)
    except ZeroDivisionError:
        print ("Zero division is not possible")
else: print ("Please enter a valid number: 1-4")
```
